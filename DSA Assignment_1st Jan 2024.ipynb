```python
#Problem 7: Swap nodes in pairs in a linked list.


class Node:
    # Constructor to initialize the node object
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    # Function to initialize head
    def __init__(self):
        self.head = None

    # Function to pairwise swap elements of a linked list using recursion
    def pairWiseSwap(self, head):
        # There must be at least two nodes in the list
        if head is not None and head.next is not None:
            # Swap the node's data with data of the next node
            head.data, head.next.data = head.next.data, head.data
            # Call pairWiseSwap() for the rest of the list
            self.pairWiseSwap(head.next.next)

    # Function to insert a new node at the beginning
    def next(self, new_data):
        new_node = Node(new_data)
        new_node.next = self.head
        self.head = new_node

    # Utility function to print the linked LinkedList
    def printList(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()

# Driver program
llist = LinkedList()
llist.next(4)
llist.next(3)
llist.next(2)
llist.next(1)

print("Linked list ")
llist.printList()

# Call pairWiseSwap() on the linked list
llist.pairWiseSwap(llist.head)

print("\nLinked list after calling pairWiseSwap()")
llist.printList()
```

```python
#Problem 8: Reverse nodes in a linked list in groups of k.

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Creating nodes
head = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)
node5 = Node(5)
node6 = Node(6)
node7 = Node(7)
node8 = Node(8)

# Connecting nodes to form a linked list
head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6
node6.next = node7
node7.next = node8

def reverse_linked_list(head):
    prev = None
    current = head

    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev

def reverse_in_groups(head, k):
    if not head or k <= 1:
        return head

    temp = None
    current = head
    count = 0

    # Count the number of nodes in the linked list
    temp_count = head
    while temp_count:
        count += 1
        temp_count = temp_count.next

    while count >= k:
        group_start = temp.next if temp else head
        group_end = group_start

        # Check if there are at least k nodes remaining
        for _ in range(k - 1):
            if group_end.next:
                group_end = group_end.next
            else:
                break

        next_group_start = group_end.next
        group_end.next = None

        # Reverse the current group and update the pointers
        if temp:
            temp.next = reverse_linked_list(group_start)
        else:
            head = reverse_linked_list(group_start)

        group_start.next = next_group_start

        # Move the pointers for the next iteration
        temp = group_start
        current = next_group_start
        count -= k

    return head

# Example usage:
# Reverse in groups of 3: 3 -> 2 -> 1 -> 6 -> 5 -> 4 -> 8 -> 7
k = 3
new_head = reverse_in_groups(head, k)

# Print the reversed linked list
while new_head:
    print(new_head.data, end=" ")
    new_head = new_head.next
```

```python
#Problem 9: Determine if a linked list is a palindrome


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Creating nodes
head = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(3)
node5 = Node(2)
node6 = Node(1)

# Connecting nodes to form a linked list
head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5
node5.next = node6

def is_palindrome(head):
    if not head or not head.next:
        return True  # An empty list or a single node is considered a palindrome

    # Step 1: Find the middle of the linked list
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    # Step 2: Reverse the second half of the linked list
    second_half = reverse_linked_list(slow)

    # Step 3: Compare the first half with the reversed second half
    while second_half:
        if head.data != second_half.data:
            return False
        head = head.next
        second_half = second_half.next

    return True

def reverse_linked_list(head):
    prev = None
    current = head

    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev

# Example usage:
# Create a palindrome linked list: 1 -> 2 -> 3 -> 3 -> 2 -> 1
result = is_palindrome(head)
print(result)  # Output: True
```

```python
#Problem 10: Rotate a linked list to the right by k places.

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Creating nodes
head = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)
node5 = Node(5)


# Connecting nodes to form a linked list
head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5


def rotate_right(head, k):
    if not head or k == 0:
        return head

    # Step 1: Find the length of the linked list
    length = 1
    current = head
    while current.next:
        length += 1
        current = current.next

    # Step 2: Calculate the effective rotation value
    k %= length

    # If k is 0 after taking the modulo, no rotation is needed
    if k == 0:
        return head

    # Step 3: Locate the new tail and head positions after rotation
    new_tail_position = length - k - 1
    new_tail = head
    for _ in range(new_tail_position):
        new_tail = new_tail.next

    new_head = new_tail.next

    # Step 4: Update the pointers to perform the rotation
    new_tail.next = None
    current.next = head

    return new_head

# Rotate the linked list to the right by 2 places
k = 2
new_head = rotate_right(head, k)

# Print the rotated linked list
while new_head:
    print(new_head.data, end=" ")
    new_head = new_head.next
```

```python
#Problem 14: Given a sorted array and a target value, return the index if the target is found. If not, return the index where it would be inserted


def search_insert_position(nums, target):
    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left  # If the target is not found, return the index where it should be inserted

# Example usage:
nums = [1, 3, 5, 6]
target = 5

result = search_insert_position(nums, target)
print("Index:", result)

 
```

```python
#Problem 15: Find the minimum element in a rotated sorted array.



def findMin(arr, low, high):
    # If the array is not rotated
    if arr[low] <= arr[high]:
        return arr[low]
 
    # Binary search
    while low <= high:
        mid = (low + high) // 2
 
        # Check if mid is the minimum element
        if arr[mid] < arr[mid-1]:
            return arr[mid]
        # If the left half is sorted, the minimum element must be in the right half
        if arr[mid] > arr[high]:
            low = mid + 1
 
        # If the right half is sorted, the minimum element must be in the left half
        else:
            high = mid - 1
 
    # If no minimum element is found, return None
    return None
# Driver program to test above functions
if __name__ == '__main__':
    arr = [4, 5, 6, 7, 0, 1, 2]
    N = len(arr)
    print("The minimum element is " + \
          str(findMin(arr, 0, N-1)))
```

```python
#Problem 16: Search for a target value in a rotated sorted array


def search(nums, target):
    if not nums:
        return -1

    left, right = 0, len(nums) - 1

    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid

        if nums[left] <= nums[mid]:
            if nums[left] <= target <= nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        else:
            if nums[mid] <= target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1

    return -1

# Example usage:
nums = [4, 5, 6, 7, 0, 1, 2]
target = 0
result = search(nums, target)
print(result)
```

```python

#Problem 17: Find the peak element in an array. A peak element is greater than its neighbors.

def find_peak_element(nums):
    left, right = 0, len(nums) - 1

    while left < right:
        mid = (left + right) // 2

        if nums[mid] > nums[mid + 1]:
            # If the element at mid is greater than its right neighbor, a peak is guaranteed on the left half
            right = mid
        else:
            # If the element at mid is less than or equal to its right neighbor, a peak is guaranteed on the right half
            left = mid + 1

    return left

# Example usage:
nums = [1, 2, 3, 1]
peak_index = find_peak_element(nums)
print(peak_index)
```

```python
#Problem 18: Given a m x n matrix where each row and column is sorted in ascending order, count the number of negative numbers.


def count_negatives(matrix):
    m, n = len(matrix), len(matrix[0])
    count = 0
    row, col = 0, n - 1

    while row < m and col >= 0:
        if matrix[row][col] < 0:
            # If the current element is negative, then all elements to the left are also negative
            count += 1
            col -= 1
        else:
            # If the current element is non-negative, move to the next row
            row += 1

    return count

# Example usage:
matrix = [
    [4, 3, 2, -1],
    [3, 2, 1, -1],
    [1, 1, -1, -2],
    [-1, -1, -2, -3]
]

result = count_negatives(matrix)
print(result)
```

```python
#Problem 19: Given a 2D matrix sorted in ascending order in each row, and the first integer of each row is greater than the last integer of the previous row, determine if a target value is present in the matrix

def search_matrix(matrix, target):
    if not matrix or not matrix[0]:
        return False

    rows, cols = len(matrix), len(matrix[0])
    left, right = 0, rows * cols - 1

    while left <= right:
        mid = (left + right) // 2
        mid_value = matrix[mid // cols][mid % cols]

        if mid_value == target:
            return True
        elif mid_value < target:
            left = mid + 1
        else:
            right = mid - 1

    return False

# Example usage:

matrix = [
    [1, 3, 5, 7],
    [10, 11, 16, 20],
    [23, 30, 34, 60]
]
target = 3

result = search_matrix(matrix, target)
print(result)  
```

```python
#Problem 20: Find Median in Two Sorted Arrays Problem: Given two sorted arrays, find the median of the combined sorted array.


def findMedianSortedArrays(nums1, nums2):
    merged = sorted(nums1 + nums2)
    length = len(merged)

    if length % 2 == 0:
        mid1 = length // 2
        mid2 = mid1 - 1
        return (merged[mid1] + merged[mid2]) / 2.0
    else:
        mid = length // 2
        return float(merged[mid])

# Example usage:
nums1 = [1, 3]
nums2 = [2]

result = findMedianSortedArrays(nums1, nums2)
print(result) 
```

```python
#Problem 21: Given a sorted character array and a target letter, find the smallest letter in the array that is  greater than the target.

def nextGreatestLetter(letters, target):
    left, right = 0, len(letters)

    while left < right:
        mid = left + (right - left) // 2

        if letters[mid] <= target:
            left = mid + 1
        else:
            right = mid

    # If 'left' is equal to the length of the array, wrap around to the beginning
    return letters[left % len(letters)]

# Example usage:
letters = ['c', 'f', 'j']
target = 'a'

result = nextGreatestLetter(letters, target)
print(result) 
```

```python
#Problem 22: Given an array with n objects colored red, white, or blue, sort them in-place so that objects of the same color are adjacent, with the colors in the order red, white, and blue.

def sortColors(nums):
    low, mid, high = 0, 0, len(nums) - 1

    while mid <= high:
        if nums[mid] == 0:
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
        elif nums[mid] == 1:
            mid += 1
        else:
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1

# Example usage:
nums = [2, 0, 2, 1, 1, 0]
sortColors(nums)
print(nums)  
```

```python
#Problem 24: Given an unsorted array, reorder it in-place such that nums[0] <= nums[1] >= nums[2] <= nums[3]...


def wiggleSort(nums):
    for i in range(len(nums) - 1):
        if i % 2 == 0 and nums[i] > nums[i + 1]:
            nums[i], nums[i + 1] = nums[i + 1], nums[i]
        elif i % 2 == 1 and nums[i] < nums[i + 1]:
            nums[i], nums[i + 1] = nums[i + 1], nums[i]

# Example usage:
nums = [3, 5, 2, 1, 6, 4]
wiggleSort(nums)
print(nums) 
```

```python
#Problem 27: Implement linear search to find the index of a target element in an array

def linear_search(arr, target):
    for i, num in enumerate(arr):
        if num == target:
            return i
    return -1  # Return -1 if the target is not found

# Example usage:
input_array = [5, 3, 8, 2, 7, 4]
target_element = 8

output_index = linear_search(input_array, target_element)
print(output_index) 
```

```python
#Problem 28 Calculate the factorial of a given number.

def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Example usage:
input_number = 5
output_factorial = factorial_iterative(input_number)
print(output_factorial)  
```

```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)

# Example usage:
input_number = 5
output_factorial = factorial_recursive(input_number)
print(output_factorial) 
```

```python
#Problem 29: Check if a given number is a prime number

import math

def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(math.sqrt(number)) + 1):
        if number % i == 0:
            return False
    return True

# Example usage:
input_number = 17
output_result = is_prime(input_number)
print(output_result) 
```

```python
#Problem 30: Generate the Fibonacci series up to a given number n.


def generate_fibonacci(n):
    fibonacci_series = [0, 1]

    while fibonacci_series[-1] + fibonacci_series[-2] <= n:
        next_fibonacci = fibonacci_series[-1] + fibonacci_series[-2]
        fibonacci_series.append(next_fibonacci)

    return fibonacci_series

# Example usage:
input_number = 8
output_fibonacci_series = generate_fibonacci(input_number)
print(output_fibonacci_series)
```

```python
def generate_fibonacci_recursive(n, series=[0, 1]):
    if series[-1] + series[-2] > n:
        return series
    else:
        next_fibonacci = series[-1] + series[-2]
        series.append(next_fibonacci)
        return generate_fibonacci_recursive(n, series)

# Example usage:
input_number = 8
output_fibonacci_series_recursive = generate_fibonacci_recursive(input_number)
print(output_fibonacci_series_recursive) 
```

```python
#Problem 31: Calculate the power of a number using recursion

def power_recursive(base, exponent):
    if exponent == 0:
        return 1
    else:
        return base * power_recursive(base, exponent - 1)

# Example usage:
base = 3
exponent = 4

output_result = power_recursive(base, exponent)
print(output_result)  
```

```python
#Problem 32: Reverse a given string.

def reverse_string(input_str):
    return input_str[::-1]

# Example usage:
input_string = "Hello, World!"
output_reversed_string = reverse_string(input_string)
print(output_reversed_string) 
```

```python
def reverse_string_iterative(input_str):
    reversed_str = ""
    for char in input_str:
        reversed_str = char + reversed_str
    return reversed_str

# Example usage:
input_string = "Hello, World!"
output_reversed_string = reverse_string_iterative(input_string)
print(output_reversed_string)
```
